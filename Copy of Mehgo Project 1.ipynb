# Weekly Activity Analysis to Measure Productivity
This notebook analyses my weekly activities to measure productivity levels in various categories.  
The data includes total hours spent per week in each category.

## Step 1-4: Create the Weekly Plan DataFrame

```python

# Step 1: Create the data
data = {
    'Category': ['Study', 'Family', 'Spiritual', 'Health', 'Hobby', 'Household'],
    'Total_Weekly_Hours': [35, 84, 7, 7, 7, 42]
}

# Step 2: Import pandas
import pandas as pd

# Step 3: Create DataFrame
df = pd.DataFrame(data)

# Step 4: Display DataFrame
df
```

##Step 5: Create a bar chart to visualise weekly activities

```python

# Step 5: Create a bar chart to visualise weekly activities

# 1. Import matplotlib
import matplotlib.pyplot as plt

# 2. Plot the data
plt.bar(df['Category'], df['Total_Weekly_Hours'])

# 3. Add title and labels
plt.title("Weekly Activities Analysis")
plt.xlabel("Category")
plt.ylabel("Total Weekly Hours (hours)")

# 4. Display the chart
plt.show()
```

## 📊 Interpretation & Recommendations

**Interpretation:**
- Family activities take the largest portion of weekly hours (84 hours), indicating a strong focus on family time.
- Household chores are the second largest (42 hours), showing significant time spent on domestic responsibilities.
- Learning gets 35 hours, which is good but still less than family and household.
- Spiritual, Health, and Hobby each only get 7 hours, showing minimal attention.

**Recommendations:**
1. Increase time for Health activities to at least 14 hours weekly to improve well-being.
2. Consider allocating more time to Hobbies to maintain work-life balance.
3. Maintain a healthy balance between Family time and personal development.
